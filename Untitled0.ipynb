```python
from google.colab import drive
drive.mount('/content/drive')

# Veri yollarını belirtin
data_dir = "/content/drive/My Drive/kaan-oral_cancer/Oral_cancer_histopathological_images"  # Veri klasörünün kök dizini
class_names = ["0", "1"]  # Sınıf adları
```

```python
import os
import cv2

# Veri setini yükleyin
# Data and Labels Collection
data = []
labels = []

for class_name in class_names:
    class_dir = os.path.join(data_dir, class_name)
    for img_name in os.listdir(class_dir):
        img_path = os.path.join(class_dir, img_name)
        img = cv2.imread(img_path)
        data.append(cv2.resize(img, (64, 64)))
        labels.append(class_name)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

# 10x10 bir alt grafik düzeni oluştur
fig, axs = plt.subplots(10, 10, figsize=(20, 20))

# Veri kümenizdeki 100 görüntüyü yükleyin (bu, sizin gerçek veri setinizden gelen görüntüler olmalıdır)
# Örneğin, data listesinde saklanan 100 tane 32x32 boyutunda görüntüler olduğunu varsayalım
# Eğer resimler farklı bir boyuta sahipse, np.random.choice kullanabilirsiniz.
indices = np.random.choice(len(data), size=100, replace=False)

# Her alt grafik için bir görüntü ata
for i, idx in enumerate(indices):
    resized_img = cv2.resize(data[idx], (32, 32))  # Görüntüyü yeniden boyutlandır
    axs[i // 10, i % 10].imshow(cv2.cvtColor(resized_img, cv2.COLOR_BGR2RGB))  # Görüntüyü ata, BGR'den RGB'ye dönüştür
    axs[i // 10, i % 10].axis('off')  # Eksenleri kapat

plt.tight_layout()  # Alt grafikler arasındaki boşluğu ayarla
plt.savefig('sample_output2.png')  # Çıktıyı kaydet
plt.show()  # Görüntüyü göster
```

```python
import matplotlib.pyplot as plt
import numpy as np
import torch
import torchvision
import torchvision.transforms as transforms

# CIFAR-10 veri setini indir ve yükle
transform = transforms.Compose([
    transforms.Resize((32, 32)),  # Görüntüleri 32x32 boyutuna yeniden boyutlandır
    transforms.ToTensor(),  # Görüntüleri PyTorch tensor formatına dönüştür
])

cifar10_dataset = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform)
cifar10_loader = torch.utils.data.DataLoader(cifar10_dataset, batch_size=100, shuffle=True)

# Veri kümesinden rastgele 100 görüntü seçin
data_iter = iter(cifar10_loader)
images, labels = next(data_iter)

# 10x10 bir alt grafik düzeni oluştur
fig, axs = plt.subplots(10, 10, figsize=(20, 20))

# Her alt grafik için bir görüntü ata
for i in range(10):
    for j in range(10):
        img = images[i * 10 + j].permute(1, 2, 0).numpy()  # Görüntüyü HWC formatına dönüştür
        axs[i, j].imshow(img)
        axs[i, j].axis('off')  # Eksenleri kapat

plt.tight_layout()  # Alt grafikler arasındaki boşluğu ayarla
plt.savefig('sample_output.png')  # Çıktıyı kaydet
plt.show()  # Görüntüyü göster
```
